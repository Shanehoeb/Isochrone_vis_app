```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "serif"

cycle = np.genfromtxt(r"C:\Users\Shane\Desktop\M2 MASH\Communicating Data & Statistics\Isochrone_app\outputs\Cycling.csv", delimiter=',')
drive = np.genfromtxt(r"C:\Users\Shane\Desktop\M2 MASH\Communicating Data & Statistics\Isochrone_app\outputs\Driving.csv", delimiter=',')
transit = np.genfromtxt(r"C:\Users\Shane\Desktop\M2 MASH\Communicating Data & Statistics\Isochrone_app\outputs\Transit.csv", delimiter=',')
walk = np.genfromtxt(r"C:\Users\Shane\Desktop\M2 MASH\Communicating Data & Statistics\Isochrone_app\outputs\Walking.csv", delimiter=',')
```

```python
time = range(5,61)
figure, ax = plt.subplots()

line1, = plt.plot(time, cycle[1:], 'r', linewidth=2)
line2, = plt.plot(time, walk[1:], 'b', linewidth=2)
line3, = plt.plot(time, drive[1:], 'orange', linewidth=2)
line4, = plt.plot(time, transit[1:], 'k', linewidth=2)
plt.yticks(range(0,801, 200))
plt.title('Area of Isochrones from Les Halles', fontsize=24)
plt.xlabel(r'Time (min)', fontsize=20)
plt.ylabel(r'Area ($km^{2}$)', fontsize=20)
plt.axis([5, 62, 0, 1000])
ax.annotate("Transit", (41,700), fontsize=12)
ax.annotate("Cycling", (43,350), fontsize=12, color='r')
ax.annotate("Driving", (52,200), fontsize=12, color='orange')
ax.annotate("Walking", (52,70), fontsize=12, color='b')

plt.show()
```
